# Linear Algebra notes

Consider equations:

### Complete, non-singular system.
a + b =  10 <br>
a + 2b = 12

a = 8; b = 2; Unique solution. <br>Hence it is a complete, non-singular system. <br>

### Redundant, singular system
a + b = 10 <br>
2a + 2b = 20

a and b can have infinite solutions. <br>Hence it is a redundant, singular system. <br>

### Contradictory, singular system. 
a + b = 10 <br>
2a + 2b = 24

a and b can have no solutions. <br>Hence it is a contradictory, singular system. <br>




The equations can be represented in matrix form. <br>
Then determinant=0 --> system is singular else non-singular. <br>

## Important 1 - Determinant is calculated as ad-bc for 2x2 matrix.

## Important 2 - Constants don't matter for singularity. 

## Important 3 - Row echelon form -
- Zero rows go at bottom
- Each row has a pivot (leftmost non-zero entry)
- Every pivot is to the right of pivots on the rows above.
- Rank of the matrix is number of total pivots. 

## Important 4 - Reduced row echelon form -
- Is in row echelon form
- Each pivot is a 1
- Any number above a pivot is 0
- Rank of a matrix is number of pivots

```python
import numpy as np
```

```python
# Above equations can be written in matrix form:

# a + b = 10 ; a + 2b = 12
A = np.array([[1, 1], [1, 2]])
B = np.array([[10], [12]])

print(np.linalg.det(A))
```

```python
# a + b = 10 ; 2a + 2b = 20
A = np.array([[1, 1], [2, 2]])
B = np.array([[10], [20]])

print(np.linalg.det(A))
```

```python
# a + b = 10 ; 2a + 2b = 24
A = np.array([[1, 1], [2, 2]])
B = np.array([[10], [24]])

print(np.linalg.det(A))
```

# Eigen values and Eigen vectors


```python
# Calculating eigen vector and values

# Define a sample 2x2 matrix
A = np.array([[2, 2], 
              [1, 3]])
eigenvalues, eigenvectors = np.linalg.eig(A)

# Print the results
print("Eigenvalues:", eigenvalues)
print("Eigenvectors:", eigenvectors)
```

## Variance and co-variance

### Mean of data


### Variance of data 


### Covariance of data


```python
data = np.array([
    [10, 12, 11, 13, 12],  # Var 1
    [5, 4, 6, 5, 5],       # Var 2
    [100, 90, 110, 95, 105]# Var 3
])

means = np.mean(data, axis=1)
variances = np.var(data, axis=1, ddof=1)

print("Means:\n", means)
print("\nVariances:\n", variances)
```

```python
cov_matrix = np.cov(data)
print("\nCovariance Matrix:\n", cov_matrix)
```
