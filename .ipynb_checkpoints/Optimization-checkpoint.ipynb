# Optimization

### Goal - To reduce mininum of f(x)
f(x) can be solved analytically but requires huge compute times and complexities as dimensions of data increase. 
The solution is to use <b> gradient descent </b>

### Gradient Descent

The derivative can be defined as $\frac{df}{dx}(x_0)$ (called a **gradient**) and step to the next point (gradient descent) can be shown using the expression:

$$x_1 = x_0 - \alpha \frac{df}{dx}(x_0),\tag{1}$$

```python
import math
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# consider function f(x) = e^x - log(x)
def f(x):
    return np.exp(x) - np.log(x)
```

```python
def dfdx_example_1(x):
    return np.exp(x) - 1/x
```

```python
x_arr = np.linspace(0.001, 3, 100)
f_arr = f(x_arr)
```

```python
plt.plot(x_arr, f_arr, label="loss function")
plt.legend()
plt.show()
```

```python
x_arr = np.linspace(0.001, 8, 100)
#g = 0.1*x**2 + cos(3*x) + 2

def g(x):
    return 0.1*x**2 + math.cos(3*x) + 2
```

```python
g_arr = g(x_arr)
```

```python
plt.plot(x_arr, g_arr, label="loss function")
plt.legend()
plt.show()
```
